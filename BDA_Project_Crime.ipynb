```python
#converting TIMESTART and TIMEEND to AM and PM time format.
def convert_to_ampm(time):
    try:
        time = float(time)
        if pd.notnull(time):
            if time < 1200:
                return f"{int(time // 100)}:{int(time % 100):02d} AM"
            elif time == 1200:
                return f"{int(time // 100)}:{int(time % 100):02d} PM"
            else:
                return f"{int(time // 100 - 12)}:{int(time % 100):02d} PM"
    except (ValueError, TypeError):
        return np.nan

# Apply conversion to 'TIMESTART' and 'TIMEEND' columns
df_combined['TIMESTART'] = df_combined['TIMESTART'].apply(convert_to_ampm)
df_combined['TIMEEND'] = df_combined['TIMEEND'].apply(convert_to_ampm)

# Display the updated DataFrame
print(df_combined)
```

```python
from pyspark.sql.functions import mean
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml import Pipeline


categorical_columns = ['CODE_DEFINED', 'LarcenyCode']
num_columns = ['Arrest']

# Replace NULL values in numerical columns with mean
num_mean = crime.select([mean(col).alias(col) for col in num_columns]).collect()[0].asDict()
crime = crime.na.fill(num_mean)

# Replace NULL values in categorical columns with a default label
categorical_default = {col: "unknown" for col in categorical_columns}
crime = crime.na.fill(categorical_default)

# Indexing, encoding, and assembling
indexer = [StringIndexer(inputCol=col, outputCol=col + "_index", handleInvalid="keep") for col in categorical_columns]

encoders = [OneHotEncoder(inputCol=col + "_index", outputCol=col + "_onehot") for col in categorical_columns]

assembler = VectorAssembler(inputCols=num_columns + [col + "_onehot" for col in categorical_columns],
                            outputCol="features", handleInvalid="keep")

# Create a pipeline
crime_pipe = Pipeline(stages=indexer + encoders + [assembler])

# Apply the pipeline to the data
crime_fe = crime_pipe.fit(crime).transform(crime)
```

```python
from pyspark.sql.functions import regexp_replace, lower
from pyspark.ml.feature import Tokenizer, HashingTF
from pyspark.ml.regression import RandomForestRegressor
from pyspark.sql.functions import col
from pyspark.sql import functions as F

# Step 1: Remove Street Numbers from Addresses
# Regular expression to match a sequence of digits followed by whitespace at the beginning of a string
crime = crime.withColumn('streetName', regexp_replace(crime['ADDRESS'], '^\d+\s', ''))

#Showing the result to verify that street numbers have been removed
#crime.select('ADDRESS', 'streetName').show(truncate=False)

# Tokenization (Split the street names into words)
tokenizer = Tokenizer(inputCol="streetName", outputCol="tokens")
df_tokens = tokenizer.transform(crime)

# Step 3: Feature Transformation (Convert street names to numeric feature vectors)
hashingTF = HashingTF(inputCol="tokens", outputCol="features", numFeatures=100)
featurizedData = hashingTF.transform(df_tokens)

# Step 4: Random Forest
rf = RandomForestRegressor(featuresCol='features', labelCol='Arrest')
model = rf.fit(featurizedData)

# Step 5: Analysis (Make predictions to see which cluster each address belongs to)
predictions_rf = model.transform(featurizedData)
predictions_rf.select('streetName', 'prediction').show(truncate=False)

highest_crime_location = predictions_rf.groupBy('ADDRESS').agg(F.sum('prediction')).orderBy(F.col('sum(prediction)').desc()).first()
highest_crime_location
```

```python
# Calculate True Positives (TP), False Positives (FP), True Negatives (TN), False Negatives (FN)
TP = predictions_rf.filter((col('Arrest') == 1) & (col('prediction') == 1)).count()
FP = predictions_rf.filter((col('Arrest') == 0) & (col('prediction') == 1)).count()
TN = predictions_rf.filter((col('Arrest') == 0) & (col('prediction') == 0)).count()
FN = predictions_rf.filter((col('Arrest') == 1) & (col('prediction') == 0)).count()

# Compute Precision, Recall, and F1-score
precision = TP / (TP + FP) if (TP + FP) != 0 else 0.0
recall = TP / (TP + FN) if (TP + FN) != 0 else 0.0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0.0

# Print the results
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1_score}")
```
